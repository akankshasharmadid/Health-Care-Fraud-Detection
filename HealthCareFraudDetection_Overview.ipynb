# Case study : HealthCare provided fraud detection Analysis

# **Overview**<BR>
Health care fraud occurs when an individual, a group of people, or a company knowingly mis-represents or mis-states something about the type, the scope, or the nature of the medical treatment or service provided, in a manner that could result in unauthorized payments being made.
<br> It has a significant impact on the healthcare system since healthcare claim fraud is a premeditated crime that necessitates the engagement of certain trustworthy parties such as doctors, lab assistants, and others. As a result, detecting such fraud requires a great deal of effort because it directly affects the credibility of the physician in question. 
<br>The issue is increasingly pressing since health-care insurance firms are losing money owing to bogus claims, and as a result, they are being obliged to raise monthly insurance premiums to compensate.
<br><BR>Healthcare fraud can take several forms, including:

*   Billing for the service which was not provided (Creating fake claims)

*   Duplicate submission of claims for the same service 
*    Charging for a more complex or expensive service than was actually provided.
* Billing for a covered service when the service actually provided was not covered.




 
It is becoming increasingly vital to focus on this issue because the number of claims is growing by the day, and if someone tries to find anomalies, it will take too much time, and the average turnaround time for claim acceptance is 30 days. As a result, working with auditors to verify claim credibility is critical.

# Bussiness Problem


1.   Healthcare Insurance companies are heavily impacted by these frauudulent claims.
2.   It's vital to correctly classify the claims
3.   We can't afford to make mistakes while classifying.
4.   Our goal is to determine whether a provider is possibly fraudulent or to calculate a likelihood score of that provider's fraudulent behaviour, as well as to determine the causes behind it in order to avoid financial loss.
5.   Important to provide justification if the claim is classified as fraud. Model Interpretability is highly important.



# Machine Learning Formulation

1.   To classify the claims as Fraudulent or valid claim. It's binary classification problem.
2.   We will use the data from Kaggle Healthcare provided fraud detection analysis.
3.   We can use supervised learning algorithms for training the model.



# Business Constraints


*   Time constraint: Don't have strict latency constraints but model should not take days to run. *italicised text*
*   Cost of Misclassification : Misclassification comes at a great cost. The number of false negatives and false positives should be minimal. If a valid claim is classified as fraudulent, it calls the claim provider's reputation into doubt, and if a fraudulent claim is classified as valid, the claim provider suffers a significant loss.
* Model Interpretability : It's very important to have interpretable model. Model should not be black box.
